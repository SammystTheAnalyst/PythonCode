# Cleaning the data

```python
#Identifying the columns with null values i.e columns that do not contain any value, whether alphabet, numeric and/or alphanumeric.

df.isnull().sum()
```

```python
#Dropping redundancy columns i.e columns that are not necessary to our analysis

Suicide_Rates_in_the_World = df.drop(['population','country-year','HDI for year'], axis=1)
```

# Relationship analysis

```python
corelation = Suicide_Rates_in_the_World.corr()
```

```python
sns.heatmap(corelation, xticklabels=corelation.columns, yticklabels=corelation.columns
           ,annot=True)
```

```python
sns.pairplot(Suicide_Rates_in_the_World)
```

```python
#Between Male and Female gender, which commit suicide more each year?

sns.relplot(x='year', y='suicides_no', hue='sex', data=Suicide_Rates_in_the_World)
```

```python
#In all the age group, which commit suicides more than the other age group? Remember the first result!

sns.relplot(x='year', y='suicides_no', hue='age', data=Suicide_Rates_in_the_World)
```
